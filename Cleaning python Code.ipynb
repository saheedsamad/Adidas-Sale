```python
import pandas as pd
import numpy as np
```

```python
Adidas= pd.read_excel('Adidas US Sales Datasets.xlsx')
Adidas
```

```python
# Drop the first 3 rows
Adidas = Adidas.iloc[4:]

# Reset the index
Adidas.reset_index(drop=True, inplace=True)

# Preview the cleaned dataset
Adidas
```

```python

# Drop the first column since it is full of NaN values
Adidas = Adidas.drop(columns=Adidas.columns[0])

# Check the resulting DataFrame
Adidas.head()
```

```python
# Rename the columns
Adidas.columns = [
    "Retailer", "Retailer_ID", "Invoice_Date", "Region", "State", 
    "City", "Product", "Price_per_Unit", "Units_Sold", "Total_Sales", 
    "Operating_Profit", "Operating_Margin", "Sales_Method"
]

# Check the renamed DataFrame
Adidas.head()
```

```python
# Count the number of columns
num_columns = Adidas.shape[1]

# Count the number of rows
num_rows = Adidas.shape[0]

# Print the results
print(f"Number of columns: {num_columns}")
print(f"Number of rows: {num_rows}")
```

```python
print(Adidas.dtypes)
```

```python
# Convert Invoice_Date to datetime
Adidas['Invoice_Date'] = pd.to_datetime(Adidas['Invoice_Date'], errors='coerce')

# Convert numeric columns to appropriate types
numeric_columns = ['Price_per_Unit', 'Units_Sold', 'Total_Sales', 
                   'Operating_Profit', 'Operating_Margin']
for col in numeric_columns:
    Adidas[col] = pd.to_numeric(Adidas[col], errors='coerce')

# Verify the data types after conversion
print(Adidas.dtypes)

```

```python
# Check for duplicate rows
duplicate_rows = Adidas[Adidas.duplicated()]

# Display duplicate rows (if any)
print("Number of duplicate rows:", len(duplicate_rows))
print(duplicate_rows)

```

```python
# Check for null values in each column
null_summary = Adidas.isnull().sum()

# Display columns with null values and their count
print(null_summary)
```

```python
# Calculate the total of the Total_Sales column
total_sales = Adidas['Total_Sales'].sum()

# Display the total sales
print(f"Total Sales: {total_sales}")
```

```python

# Save the cleaned dataset to a CSV file (optional)
Adidas.to_csv('New Cleaned_Adidas_US_Sales.csv', index=False)

print("Dataset saved successfully!")
```
